# Q1. Estimating value of Pi

We can estimate the value of Pi using Monte Carlo technique. The procedure is based on probabilities and random number generation.
Here N denotes the total number of randomly generated points. If the points generated happen to lie inside the circle 
then we store the coordinates of  the points in the array "inside".The approximate value of pi is equal to 4 times the ratio of the 
number of points lying inside the circle to the total number of random points generated.

```python
import numpy
import matplotlib.pyplot as plt
plt.style.use("default")
```

```python
N = 10
inside = []
#Storing coordinates of points which lie inside the circle in the array 'inside'
for i in range(N):
    x = numpy.random.uniform(-1, 1)
    y = numpy.random.uniform(-1, 1)
    if numpy.sqrt(x**2 + y**2) < 1:
        inside.append((x, y))
        
#Plot the randomly generated points that lie inside circle       
plt.figure(figsize=(2, 2))
plt.scatter([x[0] for x in inside], [x[1] for x in inside], marker=".", alpha=0.5,color="orange")

#Printing approximate value of pi
print("Value of pi = ",4 * len(inside)/float(N))
```

# Approximate value of Pi Vs No of Iterations


```python
N = 1000000
inside = 0
radius = 1
approximate_val = numpy.empty(N)
plt.xlabel("Number of iterations")
plt.ylabel("Approximate value of pi")

for i in range(N):
    x = numpy.random.uniform(-1, 1)
    y = numpy.random.uniform(-1, 1)
    if numpy.sqrt(x**2 + y**2) < radius**2:
        inside += 1
    approximate_val[i] = 4 * inside/float(i+1)

plt.semilogx(approximate_val,color='darkblue')

# Exact value of pi is plotted in red
plt.axhline(numpy.pi, color="red", alpha=0.5);
```

# Q2b. Probability of stock going to 102 before going to 96



```python
#importing the libraries required
import numpy as np
import matplotlib.pyplot as plt
import random

#Set the maximum value of number of runs
max_n = 100000

#Function to generate the estimated probability value list
#Returns the Number of runs required for a converged value
def get_prob_list(prob_list):

    #To store the number of runs for converged value
    number_of_runs = 1

    #To count the favourable cases
    count = 0

    #Iterating over the range 1 to max_n
    for i in range(1, max_n+1):

        #Initial value of stock price
        stock_price = 100

        #Changing the stockprice until it reaches 102 or 96
        while 96 < stock_price < 102:

            #Generating a random integer in the range 0 to 1 which shows profit or loss of 1Rs every min
            change = random.randint(0, 1)

            #Updating the value of stockprice according to the change
            if change:
                stock_price+=1
            else:
                stock_price-=1

        #Incrementing the count if it is a favourable case
        if stock_price == 102:
            count+=1

        #Calculating the probability for 'i' number of runs
        prob = count/i

        #Storing all the estimated values in a list
        prob_list.append(prob)

        #To get the converged estimated value
        # Assuming the accuracy to be 0.0001 
        # The last value of 'i' (Number of runs) for which p[i]-p[i-1] >= 0.0001 will be the point of convergence
        if i > 1 and abs(prob_list[-1]-prob_list[-2]) >= 0.0001:
            number_of_runs = i+1

        #Return the number of runs for converged value
    return number_of_runs

def main():

    #Initializing the prob_list to store all the estimated values
    prob_list = []

    #Populating the prob_list and generating point of convergence
    n = get_prob_list(prob_list)

    #Printing the number of runs for converging value and estimated probability
    print(f'Number of runs = {n}\nProbability is {prob_list[-1]}')

    #Plotting the graph for number in the range 100 to max_n
    ylist = prob_list[99:]
    xlist = np.arange(100, max_n+1, 1)
    plt.figure(num=0,dpi=120,figsize=(5,3))
    plt.plot(xlist, ylist, label="Estimated value")

    #Plotting a straight line representing the analytical value of probability
    plt.axhline(y = 2/3, color = 'r', linestyle = 'dashed',label="Actual value")

    #Labeling the graph
    plt.title("Estimated value vs Number of runs")
    plt.xlabel("Number of runs")
    plt.ylabel("Probability")
    plt.legend()
    plt.show()
    
if __name__ == "__main__":
    main()
```

# Q3. Option Pricing using Black Scholes Formula

```python
import numpy as np
from scipy.stats import norm

N = norm.cdf

S = 200           #current stock price
K = 180           #strike
T = 1/12          #time to expiration
r = 0.02          #risk free interest rate
sigma = 0.15      #implied volatility

# Call option price using Black scholes formula
def BS_CALL():
    d1 = (np.log(S/K) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * N(d1) - K * np.exp(-r*T)* N(d2)

# Put option price using Black scholes formula
def BS_PUT():
    d1 = (np.log(S/K) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma* np.sqrt(T)
    return K*np.exp(-r*T)*N(-d2) - S*N(-d1)

print("Call option Price : ",BS_CALL())
print("Put option Price : ",BS_PUT())
```

# Q4. Option pricing using Monte Carlo Simulations

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

i = 1000000             # no of different paths for simulation
S0 = 200                # current stock price
K = 180                 # strike
r = 0.02                # risk free interest rate
sigma = 0.15            # volatility
T = 1/12                # time to expiration
N = 240                 # time steps for a path
dt = T / N
discount_factor = np.exp(-r * T)

# Geometric Brownian motion path generating function
def gen_paths(S0, r, sigma, T, N, I):
    dt = float(T) / N
    paths = np.zeros((N + 1, I), np.float64)
    paths[0] = S0
    for t in range(1, N + 1):
        rand = np.random.standard_normal(I)
        paths[t] = paths[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt +
                                         sigma * np.sqrt(dt) * rand)
    return paths

# generate paths
np.random.seed(123)
paths = gen_paths(S0, r, sigma, T, N, i)

# paths[-1] gives the value of stock at expiration (St)
V = np.maximum(0, paths[-1] - K)

# The payoff will be the max of (St - K, 0)
CallPayoffAverage = np.average(np.maximum(0, paths[-1] - K))
CallPayoff = discount_factor * CallPayoffAverage

print("Call option price :",CallPayoff)
```

```python
# plotting option price vs number of runs starting from 0 to 100000
I = 100000
sim = np.zeros(I)
for x in range(0,I):
    sim[x] = discount_factor*V[0:x+1].mean()
    
plt.figure(figsize=(6, 4))
plt.title('Estimated Option Price Vs No of Runs',fontsize=15)
plt.xlabel('Iterations')
plt.ylabel('Option Price')
plt.plot(sim,color='darkblue');
```

```python
# Histogram of Value of call option depicting underlying asset prices have lognormal distribution

plt.figure(figsize=(5,3))
plt.hist(np.exp(-r * T) * V, bins=45)
plt.title('Histogram of Value of the Call Option',fontsize=15,pad = 20)
plt.axvline(np.exp(-r * T) * V.mean(), c='red')
plt.show()
```
